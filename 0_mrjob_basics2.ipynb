```python
#the line below saves the file as 0_mrjob_basics2.py
%%file 0_mrjob_basics2.py
#The simplest way to write a one-step job is to subclass MRJob and override a few methods
from mrjob.job import MRJob 
import re

#split a line at comma boundaries
WORD_RE = re.compile(r"[\w']+")
class MRWordFreqCount(MRJob):
  def mapper(self, _, title):
    fields = title.strip().split(',')         # Split title to fields using the data delimeter
    description = fields[3]                   # Select the required field
    for word in WORD_RE.findall(description):
      yield word.lower(), 1                       # Use a word as a key
  
  def combiner(self, word, counts):
    yield word, sum(counts)
  
  def reducer(self, word, counts):
    yield word, sum(counts)

if __name__ == '__main__': 
  MRWordFreqCount.run()
```
